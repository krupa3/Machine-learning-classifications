```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing
```

```python
credit=pd.read_csv('german_credit.csv')
```

```python
credit
```

```python
x=credit.loc[:, ['Credit Amount','Purpose','Account Balance']]
```

```python
x.describe()
```

```python
x=credit.iloc[:,1:31]
y=credit.iloc[:,0]
```

```python
x
```

```python
y
```

```python
minmax=preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(x).transform(x)
```

```python
from sklearn import cross_validation, neighbors
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

# Logistic Regression

```python
from sklearn.linear_model.logistic import LogisticRegression
cls =LogisticRegression(random_state =0)
lr_cls=cls.fit(x_train, y_train)
```

```python
lr_y_test =lr_cls.predict(x_test)
lr_y_test
```

```python
lr_y_train = lr_cls.predict(x_train)
lr_y_train
```

```python
from sklearn.metrics import confusion_matrix
lr_cm_test = confusion_matrix(y_test, lr_y_test)
lr_cm_test
```

```python
from sklearn.metrics import accuracy_score
lr_accu_test= accuracy_score(y_test,lr_y_test)
lr_accu_test
```

```python
lr_accu_train= accuracy_score(y_train,lr_y_train)  #train 98--rs=50  
lr_accu_train
```

```python
from sklearn.metrics import precision_score

lr_precision_test =precision_score(y_test, lr_y_test, average='weighted')  
print(lr_precision_test)
lr_precision_train =precision_score(y_train, lr_y_train, average='weighted')
print(lr_precision_train)
```

```python
from sklearn import model_selection
from sklearn.ensemble import AdaBoostClassifier
seed = 7
num_trees = 30
kfold = model_selection.KFold(n_splits=10, random_state=seed)
model = AdaBoostClassifier(n_estimators=num_trees, random_state=seed)
results = model_selection.cross_val_score(model, x_train, y_train, cv=kfold)
```

```python
results.mean()
```

```python
accuracy = 100 - results.mean()
print('Accuracy:', round(accuracy, 2), '%.')
```
